```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url = 'https://www.nst.com.my/news/nation/2019/12/545445/ron97-cost-2-sen-cheaper-ron95-and-diesel-unchanged'
# fetch content from url
page_response = requests.get(url, timeout=5)
# parse html
page_content = BeautifulSoup(page_response.content, "html.parser")
```

```
titles = page_content.select('h1')
titles = titles[0].text.strip()

titles
```

```
# Get the image
imgs = page_content.select('div.field-content img')
# Only one element in an img thus can access with array
img = imgs[0]['src']

img
```

```
# Get the text content
nodes = page_content.select('div.field-items p')
# Inspect nodes and combine all the diff p into a single string
content = ''
for node in nodes:
  content = content + node.text

print(content)
```

```
news = {
    'titles' : titles,
    'image' : img,
    'content' : content
}

news 
```
